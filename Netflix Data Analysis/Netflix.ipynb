## In which year highest number of TV shows and movies were released?

```python
#Convert date_added from object to datetime
df['date_added'] = pd.to_datetime(df['date_added'])
```

## Maximum Duration of a TV Show on Netflix

```python
df.head(3)
```

```python
df['duration'].unique()
```

```python
df[['Minutes', 'Unit']] = df['duration'].str.split(' ',expand = True)
```

```python
df[['Minutes', 'Unit']]
```

```python
df['Minutes'] = df['Minutes'].astype(int)
```

```python
df['Minutes'] .dtype
```

```python
df[(df['Minutes']== df['Minutes'].max()) &(df['type'] == 'TV Show')]
```
